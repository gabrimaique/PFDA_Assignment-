```python
# first we have to import the libraries to run the code and perform analysis
import numpy as np # library that facilitates efficient numerical operations on large quantities of data
import pandas as pd #Python package that is most widely used for data science/data analysis and machine learning tasks.
import seaborn as sns #library for making statistical graphics in Python
from scipy.stats import norm #solve scientific and mathematical problems.
import matplotlib.pyplot as plt #library for creating static, animated, and interactive visualizations
```

# Variables

## Age

I started  creating a data set to generate  randomly a set of 100 randomly generated integers in a range from 16 to 80 to represent the ages of the voters.I made the decision to fix the seed for all uses of the numpy.random package in my code for the sake of consistency and reproducibility of the results. Having generated these numbers I then plot a histogram to show the distribution of the variable across the simulated data , which you can see in the code and plot below.

Since the age variable is a continuous numerical number, it can take any value falling inside a specific range, in this case, 16 to 80 years of age (which is approximately the age range of voters in Brazil).

156 million Brazilians are required to cast their vote at the polls. Between this sizable voter population, 15% are under 25-year-olds, 40% are people in their middle age, and the remainder you can see in the graph below.


```python
np.random.seed(12345)#explicitly set a seed your code will have repeatable outputs
voters_age = ["16_to_24", "25_to_45", "45_to_60", "60_to_80"]#creates a list of 4 diferent group of ages
voters_age_distribution = np.random.choice(voters_age, 100, p=[0.15, 0.40, 0.20, 0.25 ])#generates an array of 100 numbers 
print(voters_age_distribution)
```

```python
plt.hist(voters_age_distribution, edgecolor='k')#function in pyplot module of matplotlib library is used to plot a histogram
plt.title("Distribution of ages")
```

```python
df1 = pd.DataFrame(data=voters_age_distribution, columns=["voters_age"])
df1 #pd.DataFrame organizes data into a 2-dimensional table of rows and columns
```

```python
df1.describe()#The describe() method returns description of the data in the DataFrame
```

## Gender

In Brazil, women make up 52% of the electorate while males make up 48% of the electorate.

```python
np.random.seed(12345)
gender = ['Male', 'Female'] 
voters_gender = np.random.choice(gender, 100, p=[0.48, 0.52])
print(voters_gender)
```

```python
df2 = pd.DataFrame(data=voters_gender, columns=["Gender"])
df2
```

```python
df2.describe()
```

## Region


```python
labels = ["Southeast", "Northeast", "North", "South", "Midwest"]
means = [0.40, 0.25, 0.10, 0.15, 0.10 ]
x = np.arange(len(labels))
width = 0.75

ax = plt.subplot()
rects = ax.bar(x, means, width)
ax.set_xticks(x)
ax.set_xticklabels(labels)
ax.set_title("Population Distribution")
```

```python
np.random.seed(12345)
region = ["Southeast", "Northeast", "North", "South", "Midwest"]
population_distribution = np.random.choice(region, 100, p=[0.40, 0.25, 0.10, 0.15, 0.10 ])
print(population_distribution)
```

```python
df3 = pd.DataFrame(data=population_distribution, columns=["population_distribution"])
df3
```

```python
df3.describe()
```

# Social Class

```python
labels = ["A", "B", "C", "D and E"]
means = [0.05, 0.10, 0.35, 0.50 ]
x = np.arange(len(labels))
width = 0.75

ax = plt.subplot()
rects = ax.bar(x, means, width)
ax.set_xticks(x)
ax.set_xticklabels(labels)
ax.set_title("Social Class Distribution")
```

```python
np.random.seed(12345)
social_class = ["A", "B", "C", "DE"]
social_class_distribution = np.random.choice(social_class, 100, p=[0.05, 0.10, 0.35, 0.50 ])
print(social_class_distribution)
```

```python
df4 = pd.DataFrame(data=social_class_distribution, columns=["social_class_distribution"])
df4
```

```python
df4.describe()
```

# Data Frame

```python
merged_df = pd.concat([df4, df1, df2, df3], axis=1)
pd.set_option('display.max_row', 100)
merged_df.columns = ["Social Class", "Voters Age", "Gender", "Region"]
merged_df 
```
