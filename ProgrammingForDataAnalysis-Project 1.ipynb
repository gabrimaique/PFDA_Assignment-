
The study will investigate or explore 4 variables across a dataset of 100 customer. The variables of interest are:

* Age (Continuous: 17-65)
* Gender (Categorical: Male or female)
* Region (Coninuous: 30-100)
* Candidate


```python
import pandas as pd 
import numpy as np 
import seaborn as sns 
import matplotlib.pyplot as plt 
```

```python
region = (["North", "Midwest", "Southeast", "South", "Midwest" ]) 

df = pd.DataFrame(np.random.multinomial(100,[5/25.]*5, size=1)) 

df.columns = (["North", "Midwest", "Southeast", "South","Midwest"])

df.index = (["Total"])

print (df)
```

```python
gender = (["Male", "Female"])

df = pd.DataFrame(np.random.multinomial(100,[1/2.]*2, size=1))

df.columns = (["Male", "Female"])

df.index = (["Total"])

print (df)
```

```python
voters = np.random.choice(101, 100) 
voters_ages = np.random.randint(16, 70, 100)

plt.hist(voters_ages)#plots histogram of Age
plt.xlabel("Age")#labels x axis as Age
plt.ylabel("Voters")#labels y axis as Frequency
```

```python
np.mean(voters_ages, dtype=np.integer)
```

```python
Candidate = (["Lula", "Bolsonaro"])

df = pd.DataFrame(np.random.multinomial(100,[1/2.]*2, size=1))

df.columns = (["Lula", "Bolsonaro"])

df.index = (["Total"])

print (df)
```

```python
df1 = pd.DataFrame(np.random.choice (region, 100))

df2 = pd.DataFrame(np.random.choice(gender, 100))

df3 = pd.DataFrame(voters_ages)
Candidate = (["Lula", "Bolsonaro"])

df4 = pd.DataFrame(np.random.choice (Candidate, 100))

#to merge df1, df2, df4, df4 together into one DataFrame
merged_df = pd.concat([df1, df2, df3, df4], axis=1)
pd.set_option('display.max_row', 100)
merged_df.columns = ["Region", "Gender", "Age", "Candidate"]
merged_df 
```
